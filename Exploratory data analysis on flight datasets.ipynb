# feature engineering

```python
#here the Date_of_Journey is in string format you need to convert them into three feature day months columns.
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
#drop your Date_of_Journey it's no need anymore
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
#taking only time from the arrival time not the date etc
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head(2)
```

```python
#convert your arrival time to hour and min

df['Arrival_Hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype(int)

df['Arrival_min']=df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype(int)

```

```python
#drop your Arrival_Time now

df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
#convert your dep_time to hour and min

df['Dep_Time_Hour']=df['Dep_Time'].apply(lambda x:x.split(':')[0]).astype(int)
df['Dep_Time_min']=df['Dep_Time'].apply(lambda x:x.split(':')[1]).astype(int)

df.drop('Dep_Time',axis=1,inplace=True)
```
